<i>Pre-Processing<i/>

```python
from sklearn.preprocessing import LabelEncoder
# encode values as:-
# red green blue
#  |    |    |
#  0    1    2
le=LabelEncoder()
obj_col = df.select_dtypes(include=['object']).columns
for col in obj_col:
  df[col] = le.fit_transform(df[col])
```

```python
df.head(50)
```

<i>Training data split<i/>

```python
from sklearn.model_selection import train_test_split
```

```python
train_df=df.drop(['ripeness'],axis=1)
test_df=df['ripeness']
```

```python
x_train,x_test,y_train,y_test=train_test_split(train_df,test_df,test_size=0.2,random_state=42)
# split the size in 80/20 ratio (test_size=0.2)
```

<i>Model<i/>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

<i>Accuracy and Classification Report</i>

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
y_pred = model.predict(x_test)
print("\nValidation Accurecy : ",accuracy_score(y_test,y_pred))
print("\nClassification Report :-\n",classification_report(y_test,y_pred))
```
